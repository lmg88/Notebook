```r
library(prophet)
library(readxl)
```

```r
ethereum<-read_xlsx(path = "R/currency/ethereum.xlsx")
```

```r
head(ethereum)
```

```r
ds<-ethereum$Date
y<-ethereum$Close
df<-data.frame(ds,y)
```

```r
plot(x=ds, y=y,type = "line")
```

```r
model<-prophet(df)
```

## Forecast

```r
future<-make_future_dataframe(model, periods = 100)
forecast<-predict(model, future)
```

```r
plot(model, forecast)
```

## Trend and seasonal components

```r
prophet_plot_components(model, forecast)
```
